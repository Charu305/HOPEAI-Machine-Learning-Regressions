```python
# Pandas is the one of the library in python, it is has the database related functions
import pandas as pd
```

```python
# From Pandas we have read csv function to read the dataset and saved it in the dataset variable
Dataset = pd.read_csv("Salary_Data.csv")
```

```python
Dataset # executed the Dataset variable to check the contain saved to it
```

```python
# Created a independent variable based on the input we are going to used for the model creation
independent = Dataset [["YearsExperience"]]
```

```python
# Created a dependent variable for output we are going to used for the model creation
dependent = Dataset [["Salary"]]
```

```python
# Imported the train test split function to split the dataset as training and test set using independent and dependent variable with the testsize as 30 %
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# imported the linear regression algorithm, saved to the varible regressor and we have fitted the independent's train set and dependent's Train set
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
# Showing the weight of the regressor model using the coef_ method
weight=regressor.coef_
weight
```

```python
# Showing the bias of the regressor model using the intercept_ method
bais=regressor.intercept_
bais
```

```python
#the trained model to make predictions based on new input data on X_test and the output stored in the Y_pred variable
Y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
# Calculating the accuracy using the r2 score as a evaluating metrics
r_score = r2_score(Y_test,Y_pred)
r_score
```

```python
# IMported the pickle librarry and created a filename to save the model
import pickle
filename = "finalized_linear_model.sav"
```

```python
#dump the regressor on to the filename and saved it
pickle.dump(regressor,open(filename,'wb'))
```

```python
# Loaded the saved mode and passed the input value and got the result
loaded_model=pickle.load(open("finalized_linear_model.sav",'rb'))
result=loaded_model.predict([[13]])
```
